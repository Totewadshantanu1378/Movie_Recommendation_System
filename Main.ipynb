```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("IMDB-Movie-Data.csv")
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
# the main features for model creation would be - 

# rank , title , genre , actors,year
```

```python
movies = data[['Rank','Title','Genre','Actors','Year']]
movies
```

```python
movies['tags'] = movies['Actors'] + movies['Genre']
new_data = movies.drop(columns = ['Genre' , 'Actors'])
new_data
```

```python
# now you have to do text classification

from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000, stop_words='english')

# there are stopwrods in the data 
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
new_data[new_data['Title'] == 'Prometheus'].index[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
# title index is 1 hence 
# we have to check the similarity for 1
```

```python
similarity[1]
```

```python
distance = sorted(list(enumerate(similarity[1])) , reverse=True , key=lambda vector:vector[1])
# converted into list and sorted in descending order
for i in distance[0:10]:
    print(new_data.iloc[i[0]].Title)
```

```python
# the above are the 10 movies whihchcan be recommended
```

```python
def recommend(movies):
    index = new_data[new_data['Title'] == movies].index[0]
    distance = sorted(list(enumerate(similarity[index])) , reverse=True , key=lambda vector:vector[1])
    for i in distance[0:10]:
       print(new_data.iloc[i[0]].Title) 

# perform the iteration for each movies
```

```python
recommend('Prometheus')
```
